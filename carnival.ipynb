Problem statement

Your task is to predict the selling price of the products based on the provided features.

```python
#%matplotlib notebook
import numpy as np
import pandas as pd
#import seaborn as sns
#import matplotlib.pyplot as plt
#import sklearn
```

```python
data_train=pd.read_csv('train.csv', engine='python')
data_test=pd.read_csv('test.csv', engine='python')
#data_add=pd.read_csv('addresses.csv', engine='python')
```

```python
data_train.fillna(0,inplace=True)
data_train

#print(len(list(set(list(data_train['Product_id'])))+list(set(list(data_test['Product_id'])))))
```

```python
data_test.fillna(0,inplace=True)
data_test.columns
```

```python
data_train.drop(['Customer_name'],axis=1,inplace=True)
data_test.drop(['Customer_name'],axis=1,inplace=True)
#data_train['deltad']=pd.to_datetime(data_train['instock_date'])
#data_test['deltad']=pd.to_datetime(data_test['instock_date'])
```

```python
data_train['lolc']=0
data_train.loc[data_train.Loyalty_customer=='Yes',['lolc']]=1

data_train.drop(['Loyalty_customer'],axis=1,inplace=True)
```

```python
data_test['lolc']=0
data_test.loc[data_test.Loyalty_customer=='Yes',['lolc']]=1

data_test.drop(['Loyalty_customer'],axis=1,inplace=True)
```

```python
#import datetime
#format= '%Y-%m-%d %H:%M:%S.%f'
#datetime.datetime.strptime(data_train['instock_date'].iloc[0],format)-datetime.datetime.strptime('',format)
```

```python

data_train['year']=0
for i in range(len(data_train)):
    
    if data_train['instock_date'].iloc[i][:4]=='2014':
        data_train['year'].iloc[i]=3
    elif data_train['instock_date'].iloc[i][:4]=='2015':
        data_train['year'].iloc[i]=2
    elif data_train['instock_date'].iloc[i][:4]=='2016':
        data_train['year'].iloc[i]=1

data_test['year']=0
for i in range(len(data_test)):
    
    if data_test['instock_date'].iloc[i][:4]=='2014':
        data_test['year'].iloc[i]=3
    elif data_test['instock_date'].iloc[i][:4]=='2015':
        data_test['year'].iloc[i]=2
    elif data_test['instock_date'].iloc[i][:4]=='2016':
        data_test['year'].iloc[i]=1
```

```python
data_train.drop(['instock_date'],axis=1,inplace=True)
data_test.drop(['instock_date'],axis=1,inplace=True)
```

```python
data_train.drop(['Product_id'],axis=1,inplace=True)
data_test.drop(['Product_id'],axis=1,inplace=True)
```

```python
rtrain_x=data_train.copy()
rtest=data_test.copy()
#rtrain_x=pd.get_dummies(data_train,columns=['Product_Category'])
rtrain_x.drop('Product_Category',axis=1,inplace=True)
rtrain_x.drop(['Selling_Price','year'],axis=1,inplace=True)
rtrain_y=data_train['Selling_Price']

from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3)
rtrain_xp=  poly.fit_transform(rtrain_x)
rtestpp=poly.fit_transform(rtest.drop(['year','Product_Category'],axis=1))
from sklearn.model_selection import train_test_split
xtrbb,xtebb,ytr,yte=train_test_split(rtrain_xp,rtrain_y,train_size=0.7)
'''======================================'''
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
xtr = scaler.fit_transform(xtrbb)
xte = scaler.transform(xtebb)
rtestp=scaler.transform(rtestpp)
'''==========================================='''

'''==========================================='''
from sklearn.linear_model import LinearRegression as crmode
#from sklearn.neighbors import KNeighborsRegressor
model = crmode()
model.fit(xtr,ytr)

#svm_predicted = model.predict(xte)

#knnreg = KNeighborsRegressor(n_neighbors = 150).fit(xtr, ytr)
#confusion = confusion_matrix(yte, svm_predicted)
#confusion

from sklearn.linear_model import Ridge
linridge = Ridge(alpha=0.10).fit(xtr, ytr)
'''=================================='''
linridge.score(xte,yte)

from sklearn.linear_model import Lasso

linlasso = Lasso(alpha=1, max_iter = 10000).fit(xtr, ytr)

print(model.score(xtr,ytr),linridge.score(xtr,ytr),linlasso.score(xtr,ytr))
print(model.score(xte,yte),linridge.score(xte,yte),linlasso.score(xte,yte))
```
